## DataCleaning

```python
import numpy as np
import pandas as pd

sales_data = pd.DataFrame({'name':['william','emma','sofia','markus','edward','thomas','ethan',np.nan,'arun','anika','paul']
                          , 'region':[np.nan,'North','East',np.nan,'West','West','South',np.nan,'West','East','South']
                          , 'sales':[50000,52000,np.nan,np.nan,42000,72000,49000,np.nan,67000,65000,67000]
                          , 'expenses':[42000,43000,np.nan,np.nan,38000,39000,42000,np.nan,39000,50000,45000]})
print(sales_data)
```

```python
print(sales_data.isna()) # To check if there is any null data (NaN) in the table returns boolean datatype
print(sales_data.isna().sum()) # gives the total number of null values (NaN) in the rows

# 1st method
print(sales_data.dropna())
```

```python
print(sales_data.isna()) # To check if there is any null data (NaN) in the table returns boolean datatype
print("------------------------")
print(sales_data.isna().sum()) # gives the total number of null values (NaN) in the rows
print("------------------------")

# 1st method
print(sales_data.dropna()) # it does not changes, it just shows
# sales_data = sales_data.dropna() # this changes the table
print("------------------------")


print(sales_data)
print("------------------------")

print("how Parameter")
print(sales_data.dropna(how='all'))
print("------------------------")
print("thresh parameter")
print(sales_data.dropna(thresh=3))


sales_data.dropna(subset=['sales','expenses'])
```
