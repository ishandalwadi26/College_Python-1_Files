```python
import numpy as np
time = np.arange(12)
income = np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expenses = np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expenses,color='red')
plt.fill_between(time,income,expenses,where=(income > expenses),color='green',alpha=1,label='Positive',interpolate=True)
# interpolate=True is used for filling the gaps
plt.fill_between(time,income,expenses,where=(income <= expenses),color='red',alpha=1,label='Negative',interpolate=True)
plt.xlabel('Time')
plt.ylabel('Income / Expenses Comparision')
plt.legend()
plt.show()
```

```python
import pandas as pd
import folium

dataset = pd.read_csv('PDI.csv')  
print(dataset.shape)  

# Select the first 100 rows
df = dataset.iloc[0:100]
print(df.shape)

lattiude = 37.77
longitute = -122.42

sanfran_map = folium.Map(location=[lattiude,longitute],zoom_strat=12)
sanfran_map

incidents = folium.map.FeatureGroup()
for lat,long in zip(df.Y, df.X):
    incidents.add_child(folium.CircleMarker([lat,long], radius=5, color='yellow', fill=True, fill_color='black',fill_opacity=0.6))
sanfran_map.add_child(incidents)  
```

```python
import pandas as pd
import folium

dataset = pd.read_csv('PDI.csv')  
print(dataset.shape)  

# Select the first 100 rows
df = dataset.iloc[0:100]
print(df.shape)

lattiude = 37.77
longitute = -122.42

sanfran_map = folium.Map(location=[lattiude,longitute],zoom_strat=12)
sanfran_map

incidents = folium.map.FeatureGroup()
for lat,long,labels in zip(df.Y, df.X, df.Category):
    incidents.add_child(folium.CircleMarker([lat,long], radius=5, color='yellow', fill=True, fill_color='black',fill_opacity=0.6),
                       folium.Marker([lat,long],popup=labels).add_to(sanfran_map))
sanfran_map.add_child(incidents)  
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# g = nx.DiGraph() # for directed 
g = nx.Graph() # for undirected

g.add_node(1)
g.add_nodes_from([2,3])
g.add_nodes_from(range(4,7))

g.add_edge(1,2)
g.add_edge(1,1)
g.add_edges_from([(2,3),(3,6),(5,6),(2,6),(3,1),(4,4),(4,2),(6,7)])

nx.draw_networkx(g, node_size = 850, node_color= 'g', width = 2, edge_color= 'blue')

plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

g = nx.DiGraph() 

l = [[0,1,0,1],[0,0,0,1],[0,0,0,1],[1,0,1,0]]
g.add_nodes_from([0,1,2,3])

for i in range(4):
    for j in range(4):
        if l[i][j] == 1:
            g.add_edge(i,j)
            
nx.draw(g, with_labels = True, node_size = 850)
```
